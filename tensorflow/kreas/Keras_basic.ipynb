```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
from sklearn.datasets import load_breast_cancer
from keras.optimizers import SGD
from keras.constraints import maxnorm
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.layers import Dropout

cancer = load_breast_cancer()
data = cancer.data
target = cancer.target
```

```python
target
```

```python
len(data[0,:])
```

```python
def create_model():
    #     create model
    model = Sequential()
    model.add(Dense(60, input_dim=len(data[0,:]), activation='relu', kernel_constraint=maxnorm(3)))
    model.add(Dropout(0.2))
    model.add(Dense(30, activation='relu', kernel_constraint=maxnorm(3)))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))
    model.summary()
    #     Compile model
    sgd = SGD(learning_rate=0.1, momentum=0.9)
    model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
    return model
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_model, epochs=300, batch_size=16, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=5, shuffle=True)
results = cross_val_score(pipeline, data, target, cv=kfold)
print("Hidden: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
df= pd.read_csv(dataset_path, names=column_names,
                       comment='\t',
                      sep=" ", skipinitialspace=True)

dataset =df.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
X = df.drop(['MPG',"Horsepower"],axis =1)
y = df['MPG']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 10)
```

```python
model = Sequential()
model.add(Dense(64, input_dim = len(X_train.iloc[0,:]), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
model.summary()
model.compile(loss='mse', optimizer=keras.optimizers.RMSprop(0.001), metrics=['mae', 'mse'])
```

```python
model.fit(
  X_train, y_train,
  epochs=1000, validation_split = 0.2, verbose=1)
```

```python
predictions = model.predict(X_test)
```
