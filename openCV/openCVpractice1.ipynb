```python
# Image가 들어있는 폴더의 경로를 image_path로 지정해줍니다.
image_path = "drive/My Drive/Colab Notebooks/pattern recognition/image/"
# imread(이미지 경로, parameter) parmeter = 0 이면 흑백, 1이면 color
img1 = cv.imread(image_path+"phdcomics01.png",1)
img2 = cv.imread(image_path+"phdcomics02.png",0)

# cv2_imshow()는 이미지 띄우는거
cv2_imshow(img1)
cv2_imshow(img2)

# 이미지 저장하는거 
cv.imwrite(image_path+'graycomics.png',img2)
```

```python
# Drawing Functions

# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
img = cv.line(img,(0,0),(511,511),(255,0,0),5)
# (배경,startpoint,endpoint,color(B,G,R),thickness)

# Drawing rectangle
img = cv.rectangle(img,(0,0),(510,128),(0,255,0),3)
# (배경, stratline, ednline, color, thick)

# Drawing circle
img = cv.circle(img,(447,63), 63, (0,0,255), -1)
#(배경, center, Radius, color, -1해주면 도형을 채워줌 )

# # Drawing ellipse
img = cv.ellipse(img,(256,256),(100,100),30,0,255,255,-1)
# (배경, conter point, (xradius, yradius), angle, startangle, endangle, coloer[,thickness..]

# Drawing polygon
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv.polylines(img,[pts],True,(0,255,255))


# text도 넣을 수 있어
# Adding Text to Images
font = cv.FONT_HERSHEY_SIMPLEX
# cv.putText(	img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]	) ->	img

cv.putText(img,'sangmin',(10,450), font, 3,(255,255,255),2,cv.LINE_AA)

cv2_imshow(img)
```
